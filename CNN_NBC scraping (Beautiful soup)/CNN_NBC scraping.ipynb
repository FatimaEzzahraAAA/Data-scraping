# CNN News

```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
from datetime import date
today = date.today()

d = today.strftime("%m-%d-%y")
print("date :",d)
```

```python
cnn_url = "https://edition.cnn.com/business/live-news/university-protests-palestine-{}/index.html".format(d)
cnn_url
```

```python
html = requests.get(cnn_url)
```

```python
cnn_content = soup(html.content, 'lxml')
cnn_content
```

```python
for link in cnn_content.findAll("h2"):
    print("Headlines : {}".format(link.text))
```

```python
for news in cnn_content.findAll('article', {"class": 'sc-bwzfXH sc-gwVKww kDvhCz'}):
    print(news.text)
```

# NBC_News

```python
nbc_url = 'https://www.nbcnews.com/'
```

```python
r = requests.get(nbc_url)
```

```python
b= soup (r.content, 'lxml')
```

```python
for news in b.findAll('h2'):
    print(news.text)
```

```python
links = []
for news in b.findAll('h2', {'class': 'storyline__headline'}):
    links.append(news.a['href'])
    
links
```

```python
for link in links:
    page = requests.get(link)
    content = soup(page.content)
    for news in content.findAll('div', {'class': 'article-body__content'}):
        print(news.text.strip())
```
